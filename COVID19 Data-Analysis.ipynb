```python
df.drop(['S. No.'],axis=1,inplace=True)
```

```python
df
```

```python
#adding new column(Total Cases)
df['Total Cases']=df['Total Confirmed cases (Indian National)']+df['Total Confirmed cases ( Foreign National )']
```

```python
df
```

```python
#total number of cases
total_cases_overall=df['Total Cases'].sum()
print('the total no of cases till now in India is',total_cases_overall)
```

```python
#finding the active cases
df['Active Cases']=df['Total Cases']-(df['Death']+df['Cured'])
```

```python
df
```

```python
#finding the highest number of cases
#style.background_gradient is an inbuilt feature of pandas
df.style.background_gradient(cmap='Reds')
```

```python
#groupby helps to craete a group of two columns and in the end you need to join using sum function
#and also sorting the values using descending order
#since the dataframe was actually not looking fine,i used the function 'toframe' to add frames
Total_Active_Cases=df.groupby('Name of State / UT')['Active Cases'].sum().sort_values(ascending=False).to_frame()
```
